# Handling Missing values

### Fill in the Data_value column

#### Mean Imputation:
Replace missing Data_Value entries with the mean of the available values within the same group (e.g., grouped by LocationDesc, TopicDesc, MeasureDesc, etc.).

This method is simple and can be effective if the data is symmetrically distributed.

#### Grouped Median Imputation: 
The function first fills missing values within each group defined by LocationDesc, TopicDesc, and MeasureDesc. This ensures that imputation is based on similar data points.

```python
def fill_data_value(df):
    """
    This function fills in missing values in the 'Data_Value' column using median imputation.
    The median is calculated within groups defined by 'LocationDesc', 'TopicDesc', and 'MeasureDesc'.
    
    Args:
    df (pd.DataFrame): The DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: The DataFrame with filled 'Data_Value' column.
    """
    # Group by 'LocationDesc', 'TopicDesc', and 'MeasureDesc' and fill missing values with the median
    df['Data_Value'] = df.groupby(['LocationDesc', 'TopicDesc', 'MeasureDesc'])['Data_Value'].transform(lambda x: x.fillna(x.median()))
    
    # For any remaining missing values, fill with the overall median (fallback)
    overall_median = df['Data_Value'].median()
    df['Data_Value'] = df['Data_Value'].fillna(overall_median)
    
    return df

df_filled = fill_data_value(df)

# Check if there are still missing values
missing_values = df_filled['Data_Value'].isna().sum()
print(f"Missing values in 'Data_Value' column: {missing_values}")
```

### Filling in The response column whose value is dependent on the Data_value column.

#### Function Definition (response_condition): 
This nested function takes a single value from the Data_Value column and returns the corresponding Response value based on the conditions provided:

'Frequent' if Data_Value is less than 15.

'Current' if Data_Value is between 2 and 50.

'Ever' if Data_Value is up to 100.

Application of the Function (apply): The apply function iterates over each row in the DataFrame. It checks if the Response column is missing (pd.isna(row['Response'])), and if so, it fills it with the appropriate value determined by response_condition.

#### Edge Cases: 
If a Data_Value doesn't meet any of the conditions, it will leave the Response as NaN.

```python
def fill_response(df):
    """
    This function fills in missing values in the 'Response' column based on the conditions
    applied to the 'Data_Value' column:
    - 'Frequent' if Data_Value is less than 15.
    - 'Current' if Data_Value is between 2 and 50.
    - 'Ever' if Data_Value is less than or equal to 100.

    Args:
    df (pd.DataFrame): The DataFrame containing the dataset.

    Returns:
    pd.DataFrame: The DataFrame with filled 'Response' column.
    """
    def response_condition(data_value):
        if data_value < 15:
            return 'Frequent'
        elif 2 < data_value < 50:
            return 'Current'
        elif data_value <= 100:
            return 'Ever'
        else:
            return np.nan  # Just in case there's an edge case not covered by the conditions

    # Apply the function to fill the 'Response' column where it's missing
    df['Response'] = df.apply(lambda row: response_condition(row['Data_Value']) if pd.isna(row['Response']) else row['Response'], axis=1)
    
    return df

df_filled = fill_response(df)

# Check if there are still missing values
missing_responses = df_filled['Response'].isna().sum()
print(f"Missing values in 'Response' column: {missing_responses}")
```

### Fill in the Data_Value_Std_Err Column:

Imputation Based on Derived Metrics
If the Low_Confidence_Limit and High_Confidence_Limit columns are available and you assume they represent a 95% confidence interval, you can calculate the standard error using:

Standard Error = High Confidence Limit − Low Confidence Limit / 2× 1.96

```python
def fill_std_err_derived(df):
    def calculate_std_err(row):
        if pd.isna(row['Data_Value_Std_Err']):
            if not pd.isna(row['Low_Confidence_Limit']) and not pd.isna(row['High_Confidence_Limit']):
                return (row['High_Confidence_Limit'] - row['Low_Confidence_Limit']) / (2 * 1.96)
        return row['Data_Value_Std_Err']
    
    # Apply the function to fill missing standard errors
    df['Data_Value_Std_Err'] = df.apply(calculate_std_err, axis=1)
    
    # For any remaining missing values, fill with the overall median (fallback)
    overall_median = df['Data_Value_Std_Err'].median()
    df['Data_Value_Std_Err'] = df['Data_Value_Std_Err'].fillna(overall_median)
    
    return df


df_filled_derived = fill_std_err_derived(df)

# Check if there are still missing values
missing_values = df_filled_derived['Data_Value_Std_Err'].isna().sum()
print(f"Missing values in 'Data_Value_Std_Err' column (derived): {missing_values}")
```

### Fill in the Low_Confidence_Limit column:

Most suitable approach is to calculate the lower confidence limit based on a known relationship with Data_Value. 

Typically, the lower confidence limit can be computed using the formula:

Low_Confidence_Limit = Data_Value − Margin_of_Error

The margin of error is often derived from the standard error (Data_Value_Std_Err) and a critical value from a normal distribution (often corresponding to a 95% confidence interval).

**Formula:** The function uses the formula Low_Confidence_Limit = Data_Value - 1.96 * Data_Value_Std_Err to estimate the lower bound of the confidence interval assuming a 95% confidence level. The value 1.96 is the critical value for a 95% confidence interval in a normal distribution.

**Apply Function:** The apply method is used to calculate the Low_Confidence_Limit only for rows where it is missing. The lambda function performs the calculation if Low_Confidence_Limit is NaN.

**Check Missing Values:** After filling in the missing values, the script checks if there are still any missing values in the Low_Confidence_Limit column.

**Output:** The filled DataFrame is displayed to verify that the missing values have been handled.

```python
def fill_low_confidence_limit(df):
    """
    This function fills in missing values in the 'Low_Confidence_Limit' column based on the formula:
    Low_Confidence_Limit = Data_Value - 1.96 * Data_Value_Std_Err
    assuming a 95% confidence interval.
    
    Args:
    df (pd.DataFrame): The DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: The DataFrame with filled 'Low_Confidence_Limit' column.
    """
    # Calculate the lower confidence limit for missing values
    df['Low_Confidence_Limit'] = df.apply(
        lambda row: row['Data_Value'] - 1.96 * row['Data_Value_Std_Err'] 
        if pd.isna(row['Low_Confidence_Limit']) 
        else row['Low_Confidence_Limit'], axis=1
    )
    
    return df


df_filled = fill_low_confidence_limit(df)

# Check if there are still missing values
missing_values = df_filled['Low_Confidence_Limit'].isna().sum()
print(f"Missing values in 'Low_Confidence_Limit' column: {missing_values}")
```

## Fill in the High_Confidence_Limit Column:

Imputation Based on Derived Metrics
If you have the Data_Value_Std_Err and Low_Confidence_Limit, you can use the following formula to estimate the High_Confidence_Limit:

High Confidence Limit = Data Value + 1.96 × Standard Error

Derived Metrics Method: Uses the standard error and data value to calculate the confidence limit.

```python
def fill_high_confidence_limit(df):
    """
    This function fills missing values in the 'High_Confidence_Limit' column using the formula:
    High_Confidence_Limit = Data_Value + 1.96 * Data_Value_Std_Err
    
    Args:
    df (pd.DataFrame): The DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: The DataFrame with filled 'High_Confidence_Limit' column.
    """
    # Function to calculate High Confidence Limit
    def calculate_high_conf_limit(row):
        if pd.isna(row['High_Confidence_Limit']):
            if not pd.isna(row['Data_Value']) and not pd.isna(row['Data_Value_Std_Err']):
                return row['Data_Value'] + 1.96 * row['Data_Value_Std_Err']
        return row['High_Confidence_Limit']
    
    # Apply the function to fill missing high confidence limits
    df['High_Confidence_Limit'] = df.apply(calculate_high_conf_limit, axis=1)
    
    # For any remaining missing values, fill with the overall median (fallback)
    overall_median = df['High_Confidence_Limit'].median()
    df['High_Confidence_Limit'] = df['High_Confidence_Limit'].fillna(overall_median)
    
    return df

df_filled = fill_high_confidence_limit(df)

# Check if there are still missing values
missing_values = df_filled['High_Confidence_Limit'].isna().sum()
print(f"Missing values in 'High_Confidence_Limit' column: {missing_values}")
```

## Fill in the Sample_Size column:
**Imputation Based on Similar Records:** Use the median or mean Sample_Size within groups defined by relevant categorical variables (e.g., LocationDesc, TopicDesc, MeasureDesc).

**Group-wise Imputation:** Uses the median value within similar groups to fill missing Sample_Size values.

```python
def fill_sample_size(df):
    """
    This function fills missing values in the 'Sample_Size' column using median imputation
    within groups defined by 'LocationDesc', 'TopicDesc', and 'MeasureDesc'.
    
    Args:
    df (pd.DataFrame): The DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: The DataFrame with filled 'Sample_Size' column.
    """
    # Group by 'LocationDesc', 'TopicDesc', and 'MeasureDesc' and fill missing values with the median
    df['Sample_Size'] = df.groupby(['LocationDesc', 'TopicDesc', 'MeasureDesc'])['Sample_Size'].transform(
        lambda x: x.fillna(x.median())
    )
    
    # For any remaining missing values, fill with the overall median (fallback)
    overall_median = df['Sample_Size'].median()
    df['Sample_Size'] = df['Sample_Size'].fillna(overall_median)
    
    return df

df_filled = fill_sample_size(df)

# Check if there are still missing values
missing_values = df_filled['Sample_Size'].isna().sum()
print(f"Missing values in 'Sample_Size' column: {missing_values}")
```

Filling with a Common Symbol

### Fill in the GeoLocation column: 
**Group-Based Filling:** This approach fills the missing GeoLocation values based on the most frequent GeoLocation within each LocationDesc. This assumes that the same location has consistent geographic coordinates across different records.

```python
def fill_geolocation(df):
    """
    This function fills missing values in the 'GeoLocation' column based on the most common
    GeoLocation for the given 'LocationDesc'.
    
    Args:
    df (pd.DataFrame): The DataFrame containing the dataset.
    
    Returns:
    pd.DataFrame: The DataFrame with filled 'GeoLocation' column.
    """
    # Group by 'LocationDesc' and fill missing 'GeoLocation' with the most common value
    df['GeoLocation'] = df.groupby('LocationDesc')['GeoLocation'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan))
    
    return df

# Example usage:
df_filled = fill_geolocation(df)

# Check if there are still missing values
missing_values = df_filled['GeoLocation'].isna().sum()
print(f"Missing values in 'GeoLocation' column: {missing_values}")
```

# Interactions and Transformations 

**Identify Skewed Numerical Features:** Check the distribution of numerical features to identify which ones are highly skewed.


**Apply Logarithmic Transformation:** Transform these features using the logarithm. For features that have zero or negative values, consider adding a small constant before applying the logarithm.


**Update the DataFrame:** Add the transformed features to your DataFrame.

### Explanation:
**Add a Small Constant:** We add 1 to the feature values (x + 1) to avoid taking the log of zero, which is undefined.

**Apply Log Transformation:** We use np.log() to apply the natural logarithm.

**Update DataFrame:** The transformed values are added as new columns to the DataFrame with a _Log suffix.

```python
# Identify numerical columns that might be skewed
numerical_features = ['Data_Value', 'Data_Value_Std_Err', 'Scaled_Sample_Size']

# Function to apply logarithmic transformation
def apply_log_transformation(df, column):
    # Add a small constant to avoid log(0) which is undefined
    df[column + '_Log'] = df[column].apply(lambda x: np.log(x + 1) if x >= 0 else np.nan)

# Apply logarithmic transformation to each numerical feature
for feature in numerical_features:
    apply_log_transformation(df, feature)

# Check the results
df.head()
```

## 1. K-Means Clustering

### Step 1: Train the Model 

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
import pandas as pd

# Select features for clustering
X = df[['Data_Value', 'Data_Value_Std_Err', 'Scaled_Sample_Size']].dropna()

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_clusters = kmeans.fit_predict(X_scaled)

# Add cluster labels to the DataFrame
df['KMeans_Cluster'] = kmeans_clusters
```

### Step 2: Visualize the Results

```python
# Visualize K-Means Clustering
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_clusters, cmap='viridis', alpha=0.5)
plt.title('K-Means Clustering')
plt.xlabel('Standardized Data_Value')
plt.ylabel('Standardized Data_Value_Std_Err')
plt.show()
```

### Step 3: Evaluate the Model

```python
# Evaluate K-Means Clustering
kmeans_silhouette = silhouette_score(X_scaled, kmeans_clusters)
kmeans_db_index = davies_bouldin_score(X_scaled, kmeans_clusters)

print(f"K-Means Silhouette Score: {kmeans_silhouette}")
print(f"K-Means Davies-Bouldin Index: {kmeans_db_index}")
```

## 2. Agglomerative Hierarchical Clustering

### Step 1: Train the Model

```python
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns

# Apply Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=3)
agglo_clusters = agglo.fit_predict(X_scaled)

# Add cluster labels to the DataFrame
df['Agglo_Cluster'] = agglo_clusters
```

### Step 2: Visualize the Results

```python
# Visualize Agglomerative Clustering
sns.pairplot(df, hue='Agglo_Cluster', vars=['Data_Value', 'Data_Value_Std_Err', 'Scaled_Sample_Size'])
plt.suptitle('Agglomerative Clustering', y=1.02)
plt.show()
```

### Step 3: Evaluate the Model

```python
# Evaluate Agglomerative Clustering
agglo_silhouette = silhouette_score(X_scaled, agglo_clusters)
agglo_db_index = davies_bouldin_score(X_scaled, agglo_clusters)

print(f"Agglomerative Silhouette Score: {agglo_silhouette}")
print(f"Agglomerative Davies-Bouldin Index: {agglo_db_index}")
```

## 3. DBSCAN

### Step 1: Train the Model

```python
from sklearn.cluster import DBSCAN

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_clusters = dbscan.fit_predict(X_scaled)

# Add cluster labels to the DataFrame
df['DBSCAN_Cluster'] = dbscan_clusters
```

### Step 2: Visualize the Results

```python
# Visualize DBSCAN Clustering
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_clusters, cmap='plasma', alpha=0.5)
plt.title('DBSCAN Clustering')
plt.xlabel('Standardized Data_Value')
plt.ylabel('Standardized Data_Value_Std_Err')
plt.show()
```

### Step 3: Evaluate the Model

```python
# Evaluate DBSCAN Clustering
# For DBSCAN, silhouette_score can still be calculated, but note that clusters labeled as -1 are considered noise.

# Filter out noise points
dbscan_clusters_valid = dbscan_clusters[dbscan_clusters != -1]
X_scaled_valid = X_scaled[dbscan_clusters != -1]

dbscan_silhouette = silhouette_score(X_scaled_valid, dbscan_clusters_valid)
dbscan_db_index = davies_bouldin_score(X_scaled_valid, dbscan_clusters_valid)

print(f"DBSCAN Silhouette Score: {dbscan_silhouette}")
print(f"DBSCAN Davies-Bouldin Index: {dbscan_db_index}")
```

## 4. Principal Component Analysis (PCA)

### Step 1: Train the Model

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add PCA components to DataFrame
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
```

### Step 2: Visualize the Results

```python
# Visualize PCA results
plt.figure(figsize=(10, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5)
plt.title('PCA Clustering Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

### Step 3: Evaluate the Model

```python
# Evaluate PCA
explained_variance = pca.explained_variance_ratio_

print(f"PCA Explained Variance Ratio: {explained_variance}")
```
